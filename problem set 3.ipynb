# Question 1

### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this address.

### Step 3. Assign it to a variable called users

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', delimiter='|')
```

### Step 4. Discover what is the mean age per occupation


```python
OccupationMeanAge= users[['occupation','age']].groupby('occupation').mean()
OccupationMeanAge
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
df=users[['gender','occupation']]
male= df.loc[df['gender']=='M']
GroupOccupations=df.groupby('occupation').count()
MaleOccupationCount=male.groupby('occupation').count()
MaleRatio= MaleOccupationCount/GroupOccupations
MaleRatio.sort_values(ascending=False,by='gender')
```

### Step 6. For each occupation, calculate the minimum and maximum ages


```python
df=users[['occupation','age']]
MinAge= df.groupby('occupation').min()
MaxAge= df.groupby('occupation').max()
MinAge_MaxAge=pd.concat([MaxAge,MinAge],axis=1)
MinAge_MaxAge
```

### Step 7. For each combination of occupation and sex, calculate the mean age

```python
df=users[['occupation','gender','age']]
MeanAgeOfOccupationAndSex= df.groupby(['occupation','gender']).mean('age')
MeanAgeOfOccupationAndSex
```

### Step 8. For each occupation present the percentage of women and men

```python
PercentageOfWomenAndMenInOccupation= df.groupby('occupation')['gender'].value_counts(normalize=True)*100
PercentageOfWomenAndMenInOccupation
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('us-marriages-divorces-1867-2014.csv')
df
```

```python
fig= plt.figure()
plt.rcParams["figure.figsize"] = [10, 5]
df.plot( 'Year', ['Marriages_per_1000','Divorces_per_1000'], label=['Marriages','Divorces'])
plt.title('Marriages and Divorces per capita in the U.S. between 1867 and 2014', color ='Blue', fontsize= 16)

plt.xlabel('Year', color ='Black', fontsize = 17)
plt.ylabel('Capita', color ='Black', fontsize = 17)

plt.show()
```
